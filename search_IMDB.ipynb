```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

from imdb import Cinemagoer
ia = Cinemagoer()
```

```python
# Search director name to get their id and filmography data
director = ia.search_person(input("seach person : "))[0].getID()
filmography = ia.get_person_filmography(director)

# sort the filmography data by count of their roles
sorted_dict = dict(sorted(filmography['data']['filmography'].items(), key=lambda x: len(x[1]), reverse = True))

#extract roles and count from the sorted_dict
roles = list(sorted_dict.keys())
cnt = []
for i in sorted_dict:
    cnt.append(len(filmography['data']['filmography'][i]))

# plot a bar graph based on the extracted data
fig = plt.figure(figsize = (10,10))

plt.bar(roles, cnt, color = 'maroon', width = 0.5)

plt.xlabel('roles')
plt.ylabel('count')
plt.show()
```
