```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
filmography['data']['filmography']
```

```python
roles = list(filmography['data']['filmography'].keys())
cnt = []
for i in filmography['data']['filmography']:
    cnt.append(len(filmography['data']['filmography'][i]))
```

```python
sorted_data = dict(zip(roles, cnt))
sorted_data
```

```python
sorted_data = sorted(sorted_data.items(), key=lambda x:x[1])
sorted_data
```

```python
sorted_roles = []
sorted_cnt = []

for i in range(len(sorted_data)):
    sorted_roles.append(sorted_data[i][0])
    sorted_cnt.append(sorted_data[i][1])
print(sorted_roles)
```

```python
fig = plt.figure(figsize = (10,10))

#plt.bar(range(len(sorted_data)), list(sorted_data.values()), color = 'maroon', width = 0.5)
plt.bar(sorted_roles, sorted_cnt, color = 'maroon', width = 0.5)

plt.xlabel('roles')
plt.ylabel('count')
plt.show()
```
